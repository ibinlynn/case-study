## data preprocessing

###### check point: raw_data

```python
raw_data=pd.read_csv('Absenteeism-data.csv')
```

###### check point: df

```python
df = raw_data.copy()
```

```python
df=df.drop(['ID'],axis=1)
```

###### dummy variable 'reason for absence'

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
reason_columns.head()
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

###### group 'reason for absence'

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```

```python
df.head()
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

###### check point: df_reason_mod

```python
df_reason_mod = df.copy()
```

###### extract the month value

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month Value'] = list_months
```

###### extract the weekday value

```python
def date_to_weekday(date_value):
    return date_value.weekday()
df_reason_mod['Day of the Week']=df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
```

```python
df_reason_mod.columns.values
```

```python
column_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[column_names_upd]
df_reason_mod.head()
```

## logistic regression

###### creat the target

```python
df_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(df_preprocessed['Absenteeism Time in Hours'] > 
                   df_preprocessed['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
df_preprocessed['Excessive Absenteeism'] = targets
```

```python
df_preprocessed.head()
```

```python
targets.sum() / targets.shape[0]
```

```python
df_with_targets = df_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
```

###### inputs

```python
df_with_targets.shape
```

```python
unscaled_inputs = df_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs.head()
```

###### standardize the inputs

```python
from sklearn.preprocessing import StandardScaler
```

```python
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

###### Split the data into train & test and shuffle

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, #train_size = 0.8, 
                                                                            test_size = 0.2, random_state = 20)
```

```python
print (x_train.shape, y_train.shape)
```

```python
print (x_test.shape, y_test.shape)
```

##### Logistic regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

##### Manually check the accuracy

```python
model_outputs = reg.predict(x_train)
```

```python
np.sum((model_outputs==y_train)) / model_outputs.shape[0]
```

##### Finding the intercept and coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame (columns=['Feature name'], data = feature_name)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
```

```python
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
```

```python
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending=False)
```

## testing the model

```python
reg.score(x_test,y_test)
```

```python
predicted_proba = reg.predict_proba(x_test)
predicted_proba
```
